```python
import pandas as pd
df    = pd.read_csv('/kaggle/input/glassdoor-job-reviews/glassdoor_reviews.csv', sep=',', encoding='latin-1')
print("All columns : ");
print(df.columns);
Notkeep = [d for d in df.columns if d not in ['firm', 'date_review', 'job_title','work_life_balance', 'culture_values',
       'diversity_inclusion', 'career_opp', 'comp_benefits', 'senior_mgmt','recommend', 'ceo_approv', 'outlook']]
print("\nRemoved columns : ");
print(Notkeep);
data    = df.drop(columns=Notkeep)
print("\nSelected columns : ");
print(data.columns);
data['year'] = pd.DatetimeIndex(data['date_review']).year
data = data.drop(columns=['date_review'])
```

```python
import pandas as pd

average_scores = data.groupby('firm').agg({
    'work_life_balance': 'mean',
    'culture_values': 'mean',
    'diversity_inclusion': 'mean',
    'career_opp': 'mean',
    'comp_benefits': 'mean',
    'senior_mgmt': 'mean',
    'recommend': 'mean',
    'ceo_approv': 'mean'
}).reset_index()

average_scores.to_csv('average_all_scores.csv', index=False)

print("Average scores saved to average_all_scores.csv")
```

```python
title_counts = data.groupby('firm')['job_title'].value_counts().reset_index(name='title_count')
year_counts = data.groupby('firm')['year'].value_counts().reset_index(name='year_count')

title_counts.to_csv('title_counts.csv', index=False)
year_counts.to_csv('year_counts.csv', index=False)

print("Title counts saved to title_counts.csv")
print("Year counts saved to year_counts.csv")
```
